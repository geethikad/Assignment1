###### import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df = pd.concat(
    map(pd.read_csv, ['/content/college_1.csv', '/content/college_2.csv']), ignore_index=True)
print(df)
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
exceeded_expectation = df[df['CodeKata Score']>15000]
exceeded_expectation.reset_index(drop=True)
```

```python
exceeded_expectation.to_csv("/content/Exceeded_expectations.csv",index=False)
```

###### if  10000<codekata score<15000   (Reached_expectations.csv)



```python
reached_expectation = df[(df['CodeKata Score']>10000) & (df['CodeKata Score']<15000)]
reached_expectation.reset_index(drop=True)
```

```python
reached_expectation.to_csv("/content/Reached_Expectation.csv",index=False)
```

###### if  7000<codekata score<10000   (Needs_Improvement.csv)


```python
Needs_Improvements = df[(df['CodeKata Score']>7000) & (df['CodeKata Score']<10000)]
Needs_Improvements.reset_index(drop=True)
```

```python
Needs_Improvements.to_csv("/content/Needs_Improvement.csv",index=False)
```

###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
Unsatisfactory = df[(df['CodeKata Score']<7000)]
Unsatisfactory.reset_index(drop=True)
```

```python
Unsatisfactory.to_csv("/content/Least.csv",index=False)
```
